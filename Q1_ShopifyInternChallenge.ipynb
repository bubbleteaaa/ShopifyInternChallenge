This statistical summary shows that the average order amount (AOV) is \\$3145.13. The maximum order amount and maximum total items in the dataset are $704000.00 and 2000.00 respectively. These values are far from both the 50th and 75th percentile which would suggest that the data is highly skewed as this goes against the point that the sneakers are relatively affordable.

What could be going wrong with this calculation?

This large variant in amounts seem to be outliers and could mean that a customer is purchasing large number of items that an average customer would not purchase or there is an error in how the data is being obtained. 

# Visualizing the distribution of values

The visualization of the distribution of values will help see where the outliers lie

```python
# boxplot to highlight distribution of values in the dataset

plt.boxplot(df.order_amount, showmeans=True, vert=False)
plt.show()
```

The above figure is a vertical boxplot showing a distribution of the order amount column. This shows outliers that are found in the data provided. This also higlights where the mean of the order amount column lies.

```python
# boxplot to highlight distribution of values in the dataset

plt.boxplot(df.total_items, showmeans=True, vert=False)
plt.show()
```

The above figure is a vertical boxplot showing a distribution of the total items column. This shows outliers that are found in the data provided. This also higlights where the mean of the total item column lies.

A way to evaluate this data?

Upon the discovery of outliers in the order_amount and total_items column, I would suggest trimming the data. This would involve removing and placing the data that seems to be way out of range and unrealistic in another table. 

# Trimming the dataset

```python
# places values of the dataframe in a new dataframe excluding outliers above the 98th percentile

new_df = df[df['order_amount'] < np.percentile(df['order_amount'], 98.0)]
new_df
```

```python
# prints statistical summary of the new dataframe

new_df.describe()
```

This is the new resulting dataframe after removing the unrealistic outliers in the dataset

```python
# recalculate the AOV for the new dataframe

new_AOV = new_df['order_amount'].mean()
print("The AOV is: ", new_AOV)
```
